## Sourmash heatmap

### Read and prepare files

```r
# Load sourmash CVS file
sourmash_comp_matrix = file.path(sourmash_dir, "v11.cmp.csv")
sourmash_comp_matrix = read.csv(sourmash_comp_matrix)
```

```r
# Change names
sourmash_names = colnames(sourmash_comp_matrix)

# Remove file path prefixes
prefix_1 = "X.ebio.abt3_projects.vadinCA11.data.metagenome.LLMGA.HiSeqRun83.91.bin_refine.DAS_Tool..bins_DASTool_bins."
prefix_2 = "X.ebio.abt3_projects.vadinCA11.data.V11.genomes."
sourmash_names = gsub(prefix_1, "", sourmash_names)
sourmash_names = gsub(prefix_2, "", sourmash_names)
#print(sourmash_names)

# Relabel the columns
colnames(sourmash_comp_matrix) = sourmash_names
print(colnames(sourmash_comp_matrix))

# Label the rows
rownames(sourmash_comp_matrix) = colnames(sourmash_comp_matrix)

# Transform for plotting
sourmash_comp_matrix <- as.matrix(sourmash_comp_matrix)
```

### Plot heatmap

```r
hc.rows <- hclust(dist(sourmash_comp_matrix))
hc.cols <- hclust(dist(t(sourmash_comp_matrix)))
```

```r
options(repr.plot.width=20, repr.plot.height=7)
plot(hc.rows, hang = -1)
```

```r
options(repr.plot.width=20, repr.plot.height=20)
heatmap(sourmash_comp_matrix[cutree(hc.rows,k=2)==1,], Colv=as.dendrogram(hc.cols), scale='none')
```

```r
# Get lower triangle of the correlation matrix
get_lower_tri = function(cormat){
    cormat[upper.tri(cormat)] <- NA
    return(cormat)
}
```

```r
melt_sourmash_matrix = get_lower_tri(sourmash_comp_matrix[hc.cols$order, hc.cols$order]) %>% melt()
head(melt_sourmash_matrix)
```

```r
# Plot the ANI heatmap
# Adapted from https://github.com/rprops/MetaG_analysis_workflow/wiki/13.-Compute-Average-Nucleotide-Identity-(ANI)

# read file
data.ANI <- sourmash_comp_matrix

# cluster and order matrix
row.order <- hclust(dist(data.ANI))$order # clustering
col.order <- hclust(dist(t(data.ANI)))$order
dat_new <- data.ANI[row.order, col.order] # re-order matrix accoring to clustering

# melt to dataframe
df_molten_dat <- melt(as.matrix(dat_new)) # reshape into dataframe
names(df_molten_dat)[c(1:3)] <- c("bin1", "bin2", "Jaccard")

# make plot
jaccard_heatmap <- ggplot(data = df_molten_dat,
       aes(x = bin1, y = bin2, fill = Jaccard)) + 
  geom_raster() +
  scale_fill_distiller(palette = "RdYlBu", limits=c(0.75, 1)) +
  theme(axis.text.x = element_text(angle = 90, hjust = 1)) + 
  ggtitle("Clustered ggplot heatmap")+
  #geom_text(aes(label = round(Jaccard, 2)))+
  theme(axis.title=element_text(size=16), strip.text.x=element_text(size=16),
        legend.title=element_text(size=15), legend.text=element_text(size=14),
        axis.text = element_text(size=14), title=element_text(size=20),
        strip.background=element_rect(fill=adjustcolor("lightgray",0.2))
  )
```

```r
options(repr.plot.width=25, repr.plot.height=25)
jaccard_heatmap
```

```r
# Load libraries
library(reshape2)
library(ggplot2)
library(dplyr)
```

```r
# Read table with similarities and coverages
Ndb = read.csv("/ebio/abt3_projects/vadinCA11/data/V11/drep/data_tables/Ndb.csv")
head(Ndb)

# Only Methanomethylophilus comparisons
methanomethyl_genomes = c("V11_RL001.fna", 
                         "GCA_000300255.2.fna", "GCA_002503545.1.fna", 
                         "GCA_900314325.1.fna", "GCA_900313075.1.fna", 
                         "GCA_001481295.1.fna")

methanomethyl_df = Ndb[Ndb$querry %in% methanomethyl_genomes,]
```

```r
# Summary statistics of ANI and alignment coverages
methanomethyl_df$ani %>% summary
methanomethyl_df$querry_coverage %>% summary
methanomethyl_df$ref_coverage %>% summary
```

```r
# Heatmaps of coverage and ANI
# Adapted from https://github.com/rprops/MetaG_analysis_workflow/wiki/13.-Compute-Average-Nucleotide-Identity-(ANI)
```

```r
ANI_dat <- methanomethyl_df[, c("querry", "reference", "ani")]
names(ANI_dat)[c(1:3)] <- c("bin1", "bin2", "ANI")

cast_ANI = dcast(bin1~bin2, data = ANI_dat, value.var = "ANI")
row.names(cast_ANI) = cast_ANI[, 1]
cast_ANI = cast_ANI[, -1]
# cluster and order matrix
row.order = hclust(dist(cast_ANI))$order # clustering
col.order <- hclust(dist(t(cast_ANI)))$order

species_names = c("M-methylophilus alvus Mx1201", "M-methylophilus 1R26",
                  "M-methylophilus UBA78", "M-methylophilus RUG779", 
                  "M-methylophilus hRUG898", "M-methylophilus alvus RL001")
rownames(cast_ANI) = species_names
colnames(cast_ANI) = species_names
cast_ANI <- cast_ANI[row.order, col.order] # re-order matrix accoring to clustering

# melt to dataframe
ANI_dat <- melt(as.matrix(cast_ANI)) # reshape into dataframe
names(ANI_dat)[c(1:3)] <- c("bin1", "bin2", "ANI")


# ANI Heatmap
ANI_heatmap <- ggplot(data = ANI_dat,
                      aes(x = bin1, y = bin2, fill = ANI)) + 
  geom_raster() +
  scale_fill_distiller(palette = "BuGn", limits=c(0.80, 1), direction = -1) +
  theme(axis.text.x = element_text(angle = 90, hjust = 1)) + 
  #geom_text(aes(label = round(ANI, 2))) +
  theme(axis.title=element_text(size=16), strip.text.x=element_text(size=16),
        legend.title=element_text(size=15), legend.text=element_text(size=14),
        axis.text = element_text(size=16), title=element_text(size=20),
        strip.background=element_rect(fill=adjustcolor("lightgray",0.2))
  )
```

### Dendrogram of MASH ANI order Methanomassiliiicoccales

```r
# Import Mash ANI table
Mdb = read.csv("/ebio/abt3_projects/vadinCA11/data/V11/drep/data_tables/Mdb.csv")
# Mdb is in long format, needs to be cast
Mash_cast = dcast(genome1 ~ genome2, data = Mdb, value.var = "dist")

# Add genome name column as rownames and convert to distance matrix
row.names(Mash_cast) = Mash_cast[, 1]
Mash_cast = Mash_cast[, -1]
Mash_dist = Mash_cast %>% as.dist()

# Note that the y-axis plots the distance, need to fix
Mash_hclust = Mash_dist %>% hclust(., method = "average")
```

```r
# Organize label names
label_table = read.table("/ebio/abt3_projects/vadinCA11/data/V11/anvio_output/external_genomes.txt",
                       header = T, sep = "\t")

label_table$contigs_db_path = label_table$contigs_db_path %>% 
  as.vector %>% lapply(. , function(x) basename(x) %>% strsplit(., ".db", fixed = T)) %>%
  unlist

Mash_labels = Mash_hclust$labels %>% strsplit(., "\\..*", perl =  T) %>% unlist

ord_label_table = label_table[match(Mash_labels,label_table$contigs_db_path), ]

Mash_hclust$labels = ord_label_table$name
```

```r
# Set options and plot
options(repr.plot.width = 22, repr.plot.height = 12)
Mash_hclust %>% plot(., hang = -1, y = "MASH ANI", yaxt="n")

# Invert the labels in the y-axis 
# From 100% to 0% identity
axis(2, at = seq(0,1,0.2), labels = seq(100, 0, -20))
abline(h = 0.2, lty = 2)
abline(h = 0.05, lty = 3)
```
